```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
df_tesla = pdr.get_data_yahoo('TSLA')
df_tesla.head()
```

```python
df_tesla['Open'].plot(figsize=(15,6))
```

```python
df_tesla['Open: 10 days'] = df_tesla['Open'].rolling(10, min_periods=5).mean() ## Simple moving avg
```

```python
df_tesla[['Open', 'Open: 10 days']].plot(figsize=(20,6))
```

```python
df_tesla['Open'].expanding().mean().plot(figsize=(15,6)) ## Cumulative moving avg
```
